#### This code focuses on scraping data from the following site:
* https://www.ine.gob.cl/sala-de-prensa/prensa-economia/indices-de-precios-e-inflacion/1 
* This is Chile's National Statistics office CPI bulletins

```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import pandas as pd
```

```python
#Collect all of these inflation reports for Chile
    
data = []

for x in range(1, 20):
    
    print(f'{x} of 20')
    url = f"https://www.ine.gob.cl/sala-de-prensa/prensa-economia/indices-de-precios-e-inflacion/{x}"
    print(url)
    #cookies = {"platform": "pc"}
    #Parse the main page to find all article links present
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")
    articles = soup.find_all('div', class_='noticiaListadoHome col-12 col-sm-4')

    #Loop over article link
    for article in articles:
        #print(article)
        #try:
        time.sleep(0.5)
        headline = article.get_text()
        print(headline)

        # navigate to the linked article
        link = article.find("a")
        link = link["href"]
        print(link)
        link_complete = link

        # parse the linked page
        subpage = requests.get(link_complete)
        linked_soup = BeautifulSoup(subpage.content, "html.parser")   
        # tags = linked_soup.find_all("p", attrs = {'class' : 'is-style-bio-contact'})
        # tags_alt = linked_soup.find_all("span", attrs = {'class' : 'ts-research-bio-contact'})
        
        headline = linked_soup.find("h1", attrs = {'class' : 'NoticiaSingleTitulo'}).text
        date = linked_soup.find("div", attrs = {'class' : 'fechaNoticia fechaNoticiaSingleDetalle'}).text
        summary = linked_soup.find("div", attrs = {'class' : 'NoticiaSingleSummary'}).text
        content = linked_soup.find("div", attrs = {'class' : 'NoticiaSingleContent'}).text
        
        data.append({"headline": headline,"date": date, "summary": summary,"content":content,"link":link})

        # data.append({"headline": headline,"date": date_store, "tag": tagline})
        # #except:
        # print("error occurred")

df = pd.DataFrame(data)
df.to_csv('Inf_report_DB.csv')
```
